# Introduction
This notebook uses mssparkutils to copy the contents of specific directories from one abfss path to another. 
<br>With the relevant directories specified it is intended to be used to copy all relevant mapping tables from DEV to PROD when also promoting new code in Notebooks at the same time to ensure alignment of data.

```python
# Import required libraries
from notebookutils import mssparkutils as msu
```

```python
# Define DEV and PROD lakehouse paths
dev_root_path = "INSERT ABFSS PATH HERE"
prod_root_path = "INSERT ABFSS PATH HERE"

# Define abfss paths for source and sink root directories for the copy activity
source_lakehouse_root = dev_root_path
sink_lakehouse_root = prod_root_path

# Define a list of source directories and sink directories to copy and save to (they can be different paths in the relevant lakehouses)
directories_to_copy = [
    ["budget_salesperson_uploads", "Files/reference_tables/Budget Salesperson Uploads", "Files/reference_tables/Budget Salesperson Uploads"],
    ["budget_uploads", "Files/reference_tables/Budget Uploads", "Files/reference_tables/Budget Uploads"],
    ["control_tables", "Files/reference_tables/control_tables", "Files/reference_tables/control_tables"],
    ["mapping_tables", "Files/reference_tables/mapping_tables", "Files/reference_tables/mapping_tables"]
]

# Iterate through the directories list and copy the data from source to sink
for path_name, source, sink in directories_to_copy:

    # Define abfss path for DEV Lakehouse
    source_directory = source_lakehouse_root + source

    #print(source_directory)

    # Define abfss path for PROD Lakehouse
    sink_directory = sink_lakehouse_root + sink

    try:
        # Delete destination directory & files (because copy activity cannot work if there already exists directories)
        msu.fs.rm(sink_directory, True)

    except:
        print("No diectory to delete")

    # Copy source files to destination
    msu.fs.cp(source_directory, sink_directory, True)

    print(path_name + " copied from source to sink directory!")

print("All saves complete!")
```
