# Configure Example
This allows you to pass parameters to a notebook when executed from a pipeline object so that the default lakehouse is pipeline controlled. Useful when moving from DEV to TEST to PROD to change where data is interacted with.

```python
%%configure

{
    "defaultLakehouse": {  
        "name": { 
        "parameterName": "lakehouse_name", 
        "defaultValue": ""
        },
        "id": { 
        "parameterName": "lakehouse_id", 
        "defaultValue": ""
        },
        "workspaceId": { 
        "parameterName": "lakehouse_workspace", 
        "defaultValue": ""
        }

    }

}
```
