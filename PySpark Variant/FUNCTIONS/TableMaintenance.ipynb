# Iterate through all tables and perform vacuum and optimise

```python
# Get a list of all tables in the current database
all_tables = spark.sql("SHOW TABLES").collect()

# Filter out views
tables_only = [row['tableName'] for row in all_tables if row['isTemporary'] == False]

for table in tables_only:
    spark.sql(
        f"Vacuum {table} retain 168 hours")
        
    print(f"{table} table vacuumed")

    spark.sql(
        f"Optimize {table}")

    print(f"{table} table optimized")
```
