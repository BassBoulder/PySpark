```python
from azure.identity import ClientSecretCredential
from azure.keyvault.secrets import SecretClient

# Define key variables required to access key vault and authenticate using Service Principal
client_id = "ADD ID HERE""
client_secret = "ADD SECRET HERE"
tenant_id = "ADD TENANT ID HERE"
keyvault_url = "ADD KEYVAULT URL HERE"

# Create a client secret credential
credential = ClientSecretCredential(tenant_id, client_id, client_secret)

# Create a SecretClient using the credential
secret_client = SecretClient(vault_url=keyvault_url, credential=credential)

# Retrieve secret for SQL username and password
sql_username = secret_client.get_secret("NAME OF SECRET").value
sql_password = secret_client.get_secret("NAME OF SECRET").value
```
